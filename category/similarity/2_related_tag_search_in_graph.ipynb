```python
# 프로그레스 바를 출력하는 함수
def print_progress_bar(iteration, total, prefix='Complete', suffix='', decimals=5, length=100, fill='█', print_end='\r'):
    """
    Parameters:
        iteration (int): 현재 진행된 작업 수
        total (int): 전체 작업 수
        prefix (str): 프로그레스 바 앞에 추가할 텍스트
        suffix (str): 프로그레스 바 뒤에 추가할 텍스트
        decimals (int): 백분율에서 소수점 이하 자릿수
        length (int): 프로그레스 바의 총 길이 (글자 수)
        fill (str): 프로그레스 바의 채우는 문자
        print_end (str): 프로그레스 바 출력 후 커서의 위치를 조정하는 문자열
    """
    # 현재 진행된 작업의 백분율 계산
    percent = ("{0:." + str(decimals) + "f}").format(100 * (iteration / float(total)))

    # 프로그레스 바의 채워진 길이 계산
    filled_length = int(length * iteration // total)

    # 프로그레스 바 출력
    bar = fill * filled_length + '-' * (length - filled_length)
    print(f'\r{prefix} |{bar}| {percent}% {suffix}', end=print_end)

    # 모든 작업이 끝나면 줄바꿈
    if iteration == total:
        print()
```

```python
node_dict = {node:idx for idx, node in enumerate(node_list)}
```

```python
sub_dict = {}
for cur in json_file:
    try:
        node_idx = node_dict[cur]
        sub_dict[node_idx] = []

        for node_name in json_file[cur]:
            
            
                sub_dict[node_idx].append(node_dict[node_name])
    except:
        print(cur, "is not in node_dict")
```

```python
import matplotlib.pyplot as plt

# create a graph object
G = nx.Graph()

# add some nodes to the graph
for node in node_dict.values():
    G.add_node(node, label=list(node_dict.keys())[node])

for out_edge in sub_dict:
    for in_edge in sub_dict[out_edge]:
        G.add_edge(out_edge, in_edge)
```

```python
import networkx as nx
import numpy as np

taget_graph = G
target_node = [node_dict[tag] for tag in tag_list]

total = len(target_node)
iteration = 0

num_walks = 100
walk_length = 5
walks = []
# mapping = {n: i for i, n in enumerate(taget_graph.nodes)}
# taget_graph = nx.relabel_nodes(taget_graph, mapping)

for node in target_node:
    walks_2 = []
    for i in range(num_walks):
        walk = [node]
        for j in range(walk_length):
            neighbors = list(taget_graph.neighbors(walk[-1]))
            if len(neighbors) == 0:
                break
            next_node = np.random.choice(neighbors)
            walk.append(next_node)
        walks_2.extend(walk)
    walks.append(walks_2)

    # 프로그레스바
    iteration += 1
    print_progress_bar(iteration, total)
     
```

```python
from collections import Counter

reversed_node_dict = dict(map(reversed, node_dict.items()))

resault = {}
for tag, walk_node_list in zip(tag_list, walks):
    target_tags = sorted(dict(Counter(walk_node_list)).items(), key= lambda x:x[1], reverse=True)[:10]
    resault[tag] = [reversed_node_dict[tag[0]] for tag in target_tags]
```
